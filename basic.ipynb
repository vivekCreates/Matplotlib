```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,5,10)
x
```

```python
y = x**2
```

### Plot

```python
plt.plot(x,y)
plt.title("Random Value graph")
plt.xlabel("X-label")
plt.ylabel("Y-label")
```

### Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,2)
```

### Object Oriented Way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
```

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

### Saving plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.6])
ax.plot(x,y)
plt.savefig("Basic.png")
```
